# **Univariate Analysis**

. * Plot distibution of individual predictors by No-show

```python
new_df['No-show'].value_counts()/len(df)*100
```

```python
for i, predictor in enumerate(new_df.drop(columns=['No-show', 'ScheduledDay', 'AppointmentDay'])):
    plt.figure(i)
    sns.countplot(data=new_df, x=predictor, hue='No-show')
```

```python
# Get a list of predictors excluding 'No-show', 'ScheduledDay', and 'AppointmentDay'
predictors = [col for col in new_df.columns if col not in ['No-show', 'ScheduledDay', 'AppointmentDay']]

# Loop through each predictor and create a count plot
for i, predictor in enumerate(predictors):
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    ax = sns.countplot(data=new_df, x=predictor, hue='No-show')
    plt.title(f'Count plot for {predictor} vs No-show')  # Set a title for the plot
    plt.xlabel(predictor)  # Label the x-axis
    plt.ylabel('Count')  # Label the y-axis
    plt.legend(title='No-show', loc='upper right')  # Add a legend

    # Add count annotations above each bar
    for p in ax.patches:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center')

    plt.tight_layout()  # Ensure proper spacing between subplots

plt.show()  # Display all the plots





```

```python
def uniplot(df,col,title,hue =None):

    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30


    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright')

    plt.show()
```
